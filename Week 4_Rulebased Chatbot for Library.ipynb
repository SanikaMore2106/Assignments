```python
import random

# Book database
books = {
    "python": 300,
    "data science": 500,
    "machine learning": 600,
    "ai": 700
}

# Intent patterns
intents = {
    "greeting": ["hi", "hello", "hey"],
    "book": ["book", "available", "search"],
    "fine": ["fine", "late", "penalty"],
    "time": ["time", "open", "close"],
    "membership": ["membership", "join"],
    "exit": ["exit", "bye"]
}

# Function to detect intent
def get_intent(user_input):
    user_input = user_input.lower()
    for intent, patterns in intents.items():
        for word in patterns:
            if word in user_input:
                return intent
    return "unknown"

# Fine calculation (Mathematical + Exception Handling)
def calculate_fine(days):
    try:
        days = int(days)
        if days <= 0:
            return 0
        return days * 5   # ₹5 per day
    except:
        return "Invalid input!"

# Chatbot function
def chatbot():
    print("📚 Library Assistant Chatbot")
    print("Type 'exit' to stop\n")

    while True:
        user_input = input("You: ")
        intent = get_intent(user_input)

        # EXIT
        if intent == "exit":
            print("Bot: Thank you! Visit again 😊")
            break

        # GREETING
        elif intent == "greeting":
            print("Bot:", random.choice([
                "Hello! How can I help you?",
                "Hi there! What do you need?",
                "Hey! Ask me anything about library"
            ]))

        # BOOK SEARCH
        elif intent == "book":
            book = input("Bot: Enter book name: ").lower()
            if book in books:
                print(f"Bot: Book is available ✅ Price: ₹{books[book]}")
            else:
                print("Bot: Book not available ❌")

        # FINE CALCULATION
        elif intent == "fine":
            days = input("Bot: Enter number of late days: ")
            fine = calculate_fine(days)
            print(f"Bot: Your fine is ₹{fine}")

        # LIBRARY TIME
        elif intent == "time":
            print("Bot: Library is open from 9 AM to 5 PM")

        # MEMBERSHIP
        elif intent == "membership":
            print("Bot: Membership is free for students")

        # DEFAULT
        else:
            print("Bot: Sorry, I didn't understand. Type 'help'.")

# Run chatbot
chatbot()
```
